```python
sayi = int(input("4 Basamaklı sayı bilgisi giriniz"))

binler = sayi // 1000
yuzler = (sayi // 100) % 10
onlar = (sayi // 10) % 10
birler = sayi % 10

print("Sayının Rakamlar: ", binler, yuzler, onlar, birler)

yeni_sayi = yuzler * 1000 + binler * 100 + onlar * 10 + birler

print("Yeni Sayı: ", yeni_sayi)

print("Girilen Sayı: ", sayi)
```
